```python
import requests
from bs4 import BeautifulSoup

# URL de ejemplo de pisos.com
url = "https://www.pisos.com/venta/pisos-madrid/"

# Realizar la solicitud HTTP a la página
response = requests.get(url)

# Verificar si la solicitud fue exitosa
if response.status_code == 200:
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # Encontrar los anuncios
    anuncios = soup.find_all('article', class_='listing__item')

    # Almacenar los datos
    data = []

    # Extraer información de cada anuncio
    for anuncio in anuncios:
        try:
            titulo = anuncio.find('a', class_='listing__link').text.strip()
            precio = anuncio.find('span', class_='price').text.strip()
            ubicacion = anuncio.find('span', class_='location').text.strip()
            detalles = anuncio.find('div', class_='listing__details').text.strip()
            
            data.append([titulo, precio, ubicacion, detalles])
        except AttributeError:
            continue  # Saltar si falta algún dato

    # Guardar los resultados en CSV
    import pandas as pd
    df = pd.DataFrame(data, columns=['Título', 'Precio', 'Ubicación', 'Detalles'])
    df.to_csv('pisos.csv', index=False)

    print("Datos guardados en pisos.csv")
else:
    print(f"Error al obtener la página: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de ejemplo
url = "https://www.pisos.com/venta/pisos-madrid/"

# Obtener el contenido HTML
response = requests.get(url)

# Analizar el HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Encontrar anuncios (ajusta las clases según sea necesario)
anuncios = soup.find_all('article', class_='listing__item')

data = []

# Extraer información de cada anuncio
for anuncio in anuncios:
    try:
        titulo = anuncio.find('a', class_='listing__link').text.strip()
        precio = anuncio.find('span', class_='price').text.strip()
        ubicacion = anuncio.find('span', class_='location').text.strip()
        detalles = anuncio.find('div', class_='listing__details').text.strip()
        
        data.append([titulo, precio, ubicacion, detalles])
    except AttributeError:
        continue  # Saltar si falta alguno de los datos

# Crear un DataFrame
df = pd.DataFrame(data, columns=['Título', 'Precio', 'Ubicación', 'Detalles'])

# Guardar en CSV si hay datos
if not df.empty:
    df.to_csv('pisos.csv', index=False)
    print("Datos guardados en pisos.csv")
else:
    print("No se encontraron datos.")
```
