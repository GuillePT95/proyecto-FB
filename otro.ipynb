```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Inicializar el navegador
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Abrir la página de pisos.com
driver.get("https://www.pisos.com/venta/pisos-madrid_capital_zona_urbana/")

# Esperar explícitamente a que la página cargue completamente
WebDriverWait(driver, 20).until(
    EC.presence_of_element_located((By.CLASS_NAME, 'ad-preview'))
)

# Lista para almacenar los datos extraídos
inmuebles = []

# Mantener un set para evitar duplicados
inmuebles_set = set()

# Número de páginas a recorrer (ajusta según sea necesario para cargar más resultados)
num_pages = 10  

# Recorrer varias páginas o hacer scroll para cargar más anuncios
for page in range(num_pages):
    # Esperar explícitamente a que los anuncios estén presentes
    anuncios = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, 'ad-preview'))
    )

    # Extraer información de cada anuncio
    for anuncio in anuncios:
        try:
            # Extraer el título del anuncio si está presente
            titulo = anuncio.find_element(By.CLASS_NAME, 'ad-preview__title').text if anuncio.find_element(By.CLASS_NAME, 'ad-preview__title') else ''
            
            # Extraer la ubicación del anuncio si está presente
            ubicacion = anuncio.find_element(By.CLASS_NAME, 'ad-preview__subtitle').text if anuncio.find_element(By.CLASS_NAME, 'ad-preview__subtitle') else ''
            
            # Extraer el precio si está presente
            precio = anuncio.find_element(By.CLASS_NAME, 'ad-preview__price').text if anuncio.find_element(By.CLASS_NAME, 'ad-preview__price') else ''
            
            # Extraer los detalles adicionales (baños, metros cuadrados, planta, etc.)
            detalles = anuncio.find_elements(By.CLASS_NAME, 'ad-preview__char')
            habitaciones = detalles[0].text if len(detalles) > 0 else ''
            banos = detalles[1].text if len(detalles) > 1 else ''
            metros_cuadrados = detalles[2].text if len(detalles) > 2 else ''
            planta = detalles[3].text if len(detalles) > 3 else ''

            # Usar un identificador único para evitar duplicados
            identificador = f"{titulo}-{ubicacion}-{precio}"
            if identificador not in inmuebles_set:
                inmuebles.append({
                    'Título': titulo,
                    'Ubicación': ubicacion,
                    'Precio': precio,
                    'Habitaciones': habitaciones,
                    'Baños': banos,
                    'Metros Cuadrados': metros_cuadrados,
                    'Planta': planta
                })
                inmuebles_set.add(identificador)  # Añadir al set para evitar duplicados
        except Exception as e:
            print(f"Error extrayendo información: {e}")

    # Hacer clic en el botón "Siguiente" para cargar más resultados
    try:
        # Esperar y hacer clic en el botón "Siguiente"
        siguiente_boton = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CLASS_NAME, 'pagination_next single'))
        )
        siguiente_boton.click()
        time.sleep(3)  # Esperar un poco a que la nueva página se cargue
    except Exception as e:
        print(f"No se pudo encontrar el botón de 'Siguiente' o ocurrió un error: {e}")
        break  # Salir del bucle si no hay más páginas

# Cerrar el navegador
driver.quit()

# Guardar los datos en un archivo CSV
df = pd.DataFrame(inmuebles)
df.to_csv('inmuebles_pisos.csv', index=False)

print("Datos extraídos y guardados en inmuebles_pisos.csv")
```
