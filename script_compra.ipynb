```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# Configurar el driver de Chrome
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Abrir la página de pisos.com
driver.get("https://www.pisos.com/venta/pisos-madrid_capital_zona_urbana/")

# Esperar a que la página cargue completamente
time.sleep(5)

# Lista para almacenar los datos extraídos
inmuebles = []

# Esperar explícitamente a que los anuncios estén presentes
anuncios = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, 'ad-preview'))
)

# Extraer información de cada anuncio
for anuncio in anuncios:
    try:
        # Extraer título, precio, y detalles del inmueble
        titulo = anuncio.find_element(By.CLASS_NAME, 'location').text
        precio = anuncio.find_element(By.CLASS_NAME, 'price').text
        detalles = anuncio.find_element(By.CLASS_NAME, 'features').text

        # Guardar los datos
        inmuebles.append({
            'Título': titulo,
            'Precio': precio,
            'Detalles': detalles
        })
    except Exception as e:
        print(f"Error extrayendo información: {e}")

# Imprimir los inmuebles extraídos
for inmueble in inmuebles:
    print(inmueble)

# Cerrar el navegador
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# Inicializar el navegador
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Abrir la página de pisos.com
driver.get("https://www.pisos.com/venta/pisos-madrid_capital_zona_urbana/")

# Esperar a que la página cargue completamente
time.sleep(5)

# Lista para almacenar los datos extraídos
inmuebles = []

# Número de scrolls a realizar (ajusta según sea necesario para cargar más resultados)
num_scrolls = 30  

# Mantener un set para evitar duplicados
inmuebles_set = set()

# Recorrer varias páginas o hacer scroll para cargar más anuncios
for i in range(num_scrolls):
    # Esperar explícitamente a que los anuncios estén presentes
    anuncios = WebDriverWait(driver, 20).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, 'ad-preview'))
    )
    
    # Extraer información de cada anuncio
    for anuncio in anuncios:
        try:
            # Extraer el título del anuncio si está presente
            try:
                titulo = anuncio.find_element(By.CLASS_NAME, 'ad-preview__title').text
            except:
                titulo = ''  # Dejar en blanco si no se encuentra el título
            
            # Extraer la ubicación del anuncio si está presente
            try:
                ubicacion = anuncio.find_element(By.CLASS_NAME, 'ad-preview__subtitle').text
            except:
                ubicacion = ''  # Dejar en blanco si no se encuentra la ubicación
            
            # Extraer el precio si está presente
            try:
                precio = anuncio.find_element(By.CLASS_NAME, 'ad-preview__price').text
            except:
                precio = ''  # Dejar en blanco si no se encuentra el precio
            
            # Extraer los detalles adicionales (baños, metros cuadrados, planta, etc.)
            try:
                detalles = anuncio.find_elements(By.CLASS_NAME, 'ad-preview__char')
                habitaciones = detalles[0].text if len(detalles) > 0 else ''
                banos = detalles[1].text if len(detalles) > 1 else ''
                metros_cuadrados = detalles[2].text if len(detalles) > 2 else ''
                planta = detalles[3].text if len(detalles) > 3 else ''
            except:
                habitaciones = banos = metros_cuadrados = planta = ''  # Si no hay detalles, dejarlos en blanco

            # Usar un identificador único para evitar duplicados
            identificador = f"{titulo}-{ubicacion}-{precio}"
            if identificador not in inmuebles_set:
                inmuebles.append({
                    'Título': titulo,
                    'Ubicación': ubicacion,
                    'Precio': precio,
                    'Habitaciones': habitaciones,
                    'Baños': banos,
                    'Metros Cuadrados': metros_cuadrados,
                    'Planta': planta
                })
                inmuebles_set.add(identificador)  # Añadir al set para evitar duplicados
        except Exception as e:
            print(f"Error extrayendo información: {e}")

    # Hacer scroll hasta el final de la página para cargar más anuncios
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)  # Esperar para que se carguen nuevos anuncios

# Cerrar el navegador
driver.quit()

# Guardar los datos en un archivo CSV
df = pd.DataFrame(inmuebles)
df.to_csv('inmuebles3_pisos.csv', index=False)

print("Datos extraídos y guardados en inmuebles3_pisos.csv")
```
